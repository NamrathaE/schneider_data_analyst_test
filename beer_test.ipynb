#**TOP BEERS TO RECOMMEND TO SOMEONE**

```python
#Sorting the table to view beers with highest overall review value
df.sort_values(by=['review_overall'], ascending=False)
```

```python
# Number of reviews for each Beer
print(df['beer_name'].value_counts())
```

```python
# Adding a column for No. of Reviews
df['no_of_reviews'] = df.groupby('beer_name')['beer_name'].transform('size')
```

#Calculating average of overall review value for each beer

```python
#adding a column for average of overall reviews for each beer
 df['average_overall_review']= df.groupby('beer_name')['review_overall'].transform('mean')
```

```python
df.head(10)
```

#Sorting the List by average_overall_review for Beers with over 1000 reviews

```python
df[df['no_of_reviews'] > 1000].sort_values(by=['average_overall_review'], ascending=False).drop_duplicates('beer_name').head(20)
```

# **I enjoy a beer which aroma and appearance matches the beer style. What beer should I buy?**

#Calculating average of reviews on aroma

```python
df['average_aroma_review'] = df.groupby('beer_name')['review_aroma'].transform('mean')
```

```python
df.head(5)
```

```python
# Sorting in descending order by average aroma review
df.sort_values(by=['average_aroma_review'], ascending=False).drop_duplicates('beer_name').head(5)
```

#Calculating average of reviews on appearance

```python
df['average_appearance_review'] = df.groupby('beer_name')['review_appearance'].transform('mean')
```

```python
df.head(5)
```

```python
# Sorting in descending order by average appearance review
df.sort_values(by=['average_appearance_review'], ascending=False).drop_duplicates('beer_name').head(5)
```

#Aggregating Columns

```python
df.agg({'average_aroma_review': 'mean', 'average_appearance_review': 'mean'})
```
